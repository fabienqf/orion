## Gather File Info

```python
import json
import os
import pandas as pd
```

Path to the raw image library downloaded from DropBox.

```python
path_imglib = os.path.join('dropbox', 'images')
path_data = 'data'
```

```python
jsons = [os.path.join(path_imglib, x, 'metadata.json') for x in os.listdir(path_imglib)]
```

```python
metadata = [json.load(open(x)) for x in jsons]
```

```python
selected_tags = ['', '', '',
                 '', '', '',
                 '']
```

```python
def select_single(values, selected, empty='-'):
    ret = list(set(values) & set(selected))
    return ret[0] if len(ret) > 0 else empty
```

```python
imglib_info = pd.DataFrame({
    'id': [metadata[i]['id'] for i in range(len(metadata))],
    'name': [metadata[i]['name'] for i in range(len(metadata))],
    'subject': [select_single(metadata[i]['tags'], selected_tags) for i in range(len(metadata))]
})
```

Save the library metadata to a CSV file.

```python
imglib_info.to_csv(os.path.join(path_data, 'imglib_info.csv'))
```

## Reorganise Files

```python
for i in range(imglib_info.shape[0]):
    os.system('cp '
             + os.path.join(path_imglib, imglib_info.id[i] + '.info', imglib_info.name[i] + '.jpg') + ' '
             + os.path.join('data', 'train', imglib_info.id[i] + '.jpg'))
```
